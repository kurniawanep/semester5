```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
g = sns.relplot(x='Apparent Temperature (C)', y='Humidity', data=df, kind='scatter', color='indigo')
g.fig.suptitle('Apparent Temperature (C) vs. Humidity, y=1.1')
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.title("Apparent Temperature and Humidity w.r.t year")
plt.plot(df["Humidity"], label='Humidity', color='orange', marker='.')
plt.plot(df['Apparent Temperature (C)'], label='Apparent Temperature (C)', color='indigo', marker='.')
plt.xlabel("Year", fontsize=10)
plt.legend(loc=4, fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(16,6))
g = sns.relplot(x='Apparent Temperature (C)', y='Humidity', color='purple', hue='Summary', data=df, col='Summary', col_wrap=3)
g.fig.suptitle("aparent Temperature (C) vs. Humidity w.r.t Summary", y=1.005)
plt.show()
```
